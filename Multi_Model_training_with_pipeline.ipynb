```python
x=df.drop('time',axis=1)
y=df['time']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.pipeline import Pipeline  # defines sequence of data transformation
```

```python
from sklearn.compose import ColumnTransformer  # groups pipeline steps for all columns
```

```python
cat_cols=['sex','smoker','day']
num_cols=['total_bill','tip','size']
```

```python
num_pipeline=Pipeline(steps=[('imputation',SimpleImputer(strategy='median')),
                ('scaling',StandardScaler())])
```

```python
cat_pipeline=Pipeline(steps=[('imputation',SimpleImputer(strategy='most_frequent')),
                             ('encoding',OneHotEncoder())])
```

```python
preprocessor=ColumnTransformer([('num_pipeline',num_pipeline,num_cols),
                   ('cat_pipeline',cat_pipeline,cat_cols)])
```

```python
x_train=preprocessor.fit_transform(x_train)
```

```python
x_test=preprocessor.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
models={'svc':SVC(),
        'dtc':DecisionTreeClassifier(),
        'lr':LogisticRegression()}
```

```python
from sklearn.metrics import accuracy_score
```

```python
def model_train_eval(x_train,x_test,y_train,y_test,models):
  evaluation={}
  for i in range(len(models)):
    model=list(models.values())[i]
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    model_score=accuracy_score(y_test,y_pred)
    evaluation[list(models.values())[i]]=model_score
  return evaluation
```

```python
model_train_eval(x_train,x_test,y_train,y_test,models)
```
