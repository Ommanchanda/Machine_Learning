```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
x,y=make_classification(n_samples=1000,n_features=20,n_classes=2,random_state=1)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
lr=LogisticRegression()
dt=DecisionTreeClassifier()
sv=SVC()
gn=GaussianNB()
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
ensemble_clf=VotingClassifier(estimators=[('logistic_reg',lr),('decision_tree_clf',dt),('support_vector_clf',sv),('gaussian',gn)],voting='hard')
```

```python
ensemble_clf.fit(x_train,y_train)
```

```python
y_pred=ensemble_clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

**ENSEMBLE REGRESSOR**

```python
from sklearn.datasets import make_regression
```

```python
x,y=make_regression(n_samples=1000,n_features=10,noise=0.1,random_state=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lin_reg=LinearRegression()
dtr=DecisionTreeRegressor()
svr=SVR(kernel='linear')
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
ens_reg=VotingRegressor(estimators=[('lr',lin_reg),('dtr',dtr),('svr',svr)])
```

```python
ens_reg.fit(x_train,y_train)
```

```python
y_pred=ens_reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
